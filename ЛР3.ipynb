```python
import requests
from bs4 import BeautifulSoup

#Список городов и URL-адресов для каждого города
cities = { 
    'Пермский край': 'https://59.ru/text/?rubric=food', 
    'Москва': 'https://msk1.ru/text/?rubric=food', 
    'Башкортостан': 'https://ufa1.ru/text/?rubric=food' 
}

#Список блюд и соответствующих им кухонь
dishes = { 
    'Суши': 'азиатская', 
    'Похлебка': 'татарская', 
    'Борщ': 'русская',
    'Хачапури': 'грузинская',
    'Посикунчики': 'коми - пермятская', 
    'Пицца': 'итальянская' 
}

#Функция для получения данных о блюдах в городе
def get_city_dishes(city_url): 
    response = requests.get(city_url) 
    soup = BeautifulSoup(response.text, 'html.parser')
    city_dishes = []

    for dish in dishes: 
        if city_url == 'https://59.ru/text/?rubric=food':
            mentions = len(soup.select('a.ct-item-href:contains("{}")'.format(dish))) 
        elif city_url == 'https://msk1.ru/text/?rubric=food': 
            mentions = len(soup.select('a.s-i-news-box__heading-link:contains("{}")'.format(dish))) 
        elif city_url == 'https://ufa1.ru/text/?rubric=food': 
            mentions = len(soup.select('a.news__link:contains("{}")'.format(dish)))

        city_dishes.append((dish, mentions))

    return city_dishes

#Словарь для хранения данных о блюдах в каждом городе
city_dishes_data = {}

#Получение данных о блюдах для каждого города
for city, url in cities.items(): 
    city_dishes_data[city] = get_city_dishes(url)

#Вывод данных о блюдах в каждом городе
for city, dishes_data in city_dishes_data.items(): 
    print(f"{city}:")
    for dish, mentions in dishes_data: 
        print(f"{dish}: {mentions} mentions")


import matplotlib.pyplot as plt

# Создание словаря для хранения количества упоминаний каждого блюда в каждом городе
mentions_data = {}

# Заполнение словаря данными о блюдах в каждом городе
for city, dishes_data in city_dishes_data.items():
    mentions_data[city] = [mentions for dish, mentions in dishes_data]

# Создание графика
fig, axs = plt.subplots(len(dishes), len(cities))
fig.tight_layout()

# Вывод графиков для каждого блюда
for i, dish in enumerate(dishes.keys()):
    for j, city in enumerate(cities.keys()):
        axs[i, j].bar(0, mentions_data[city][i])
        axs[i, j].set_xticks([])
        axs[i, j].set_yticks([])

        if i == 0:
            axs[i, j].set_title(city)
        if j == 0:
            axs[i, j].set_ylabel(dish)

# Отображение графика
plt.show()
```
