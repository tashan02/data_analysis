```python
###импорт необходимых библиотек
import requests
import youtube_dl
import re
import os

def download_videos(urls):
    for url in urls:
        if 'imgur.com' in url:
            parse_imgur_video(url)
        elif 'redd.it' in url:
            parse_reddit_video(url)
        elif 'youtube.com' in url or 'youtu.be' in url:
            parse_youtube_video(url)
        elif 'giphy.com' in url:
            parse_giphy_video(url)
        elif 'tenor.com' in url:
            parse_tenor_video(url)
        elif 'tiktok.com' in url:
            parse_tiktok_video(url)
        elif 'rutube.ru' in url:
            parse_rutube_video(url)
        elif 'twitch.tv' in url:
            parse_twitch_video(url)
        elif 'vimeo.com' in url:
            parse_vimeo_video(url)
        else:
            print(f"Unknown video service for URL: {url}")

def parse_imgur_video(url):
    #Извлечение идентификатора видео из URL 
    video_id = re.search(r'(?<=imgur.com/)[^/?#]+', url).group(0)
    
    #Создание прямого  URL - адреса видео 
    video_url = f"https://i.imgur.com/{video_id}.mp4"

    #Создание директории для видео
    directory = f"./{video_id}"
    os.mkdir(directory)

    #Скачивание и сохранение видео
    download_video(video_url, f"{directory}/{video_id}.mp4")

def parse_reddit_video(url):
    #Извлечение идентификатора видео из URL 
    video_id = re.search(r"(?<=redd.it/)[^/?#]+", url).group(0)
    
    #Создание прямого  URL - адреса видео 
    video_url = f"https://v.redd.it/{video_id}/DASH_480"

    #Создание директории для видео
    directory = f"./{video_id}"
    os.mkdir(directory)
    #Скачивание и сохранение видео
    download_video(video_url, f"{directory}/{video_id}.mp4")

def parse_youtube_video(url):
    ydl_opts = {}
    with youtube_dl.YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])

def parse_giphy_video(url):
    video_id = re.search(r"(?<=giphy.com/)[^/?#]+", url).group(0)
    video_url = f"https://giphy.com/gifs/{video_id}/html5"
    directory = f"./{video_id}"
    os.mkdir(directory)
    download_video(video_url, f"{directory}/{video_id}.mp4")

def parse_tenor_video(url):
    video_id = re.search(r"(?<=tenor.com/)[^/?#]+", url).group(0)
    video_url = f"https://tenor.com/view/{video_id}"
    directory = f"./{video_id}"
    os.mkdir(directory)
    download_video(video_url, f"{directory}/{video_id}.mp4")

def parse_tiktok_video(url):
    video_id = re.search(r"(?<=tiktok.com/)[^/?#]+", url).group(0)
    video_url = f"https://www.tiktok.com/{video_id}"
    directory = f"./{video_id}"
    os.mkdir(directory)
    download_video(video_url, f"{directory}/{video_id}.mp4")

def parse_rutube_video(url):
    video_id = re.search(r"(?<=rutube.ru/video/)[^/?#]+", url).group(0)
    video_url = f"https://rutube.ru/play/embed/{video_id}"
    directory = f"./{video_id}"
    os.mkdir(directory)
    download_video(video_url, f"{directory}/{video_id}.mp4")

def parse_twitch_video(url):
    video_id = re.search(r"(?<=twitch.tv/)[^/?#]+", url).group(0)
    video_url = f"https://www.twitch.tv/videos/{video_id}"
    directory = f"./{video_id}"
    os.mkdir(directory)
    download_video(video_url, f"{directory}/{video_id}.mp4")

def parse_vimeo_video(url):
    video_id = re.search(r"(?<=vimeo.com/)[^/?#]+", url).group(0)
    
    video_url = f"https://vimeo.com/{video_id}"


    directory = f"./{video_id}"
    os.mkdir(directory)

    download_video(video_url, f"{directory}/{video_id}.mp4")

def download_video(url, file_path):
    response = requests.get(url, stream=True)
    with open(file_path, 'wb') as file:
        for chunk in response.iter_content(chunk_size=1024):
            file.write(chunk)

def download_videos(urls):
    for url in urls:
        if 'imgur.com' in url:
            parse_imgur_video(url)
        elif 'redd.it' in url:
            parse_reddit_video(url)
        elif 'youtube.com' in url or 'youtu.be' in url:
            parse_youtube_video(url)
        elif 'giphy.com' in url:
            parse_giphy_video(url)
        elif 'tenor.com' in url:
            parse_tenor_video(url)
        elif 'tiktok.com' in url:
            parse_tiktok_video(url)
        elif 'rutube.ru' in url:
            parse_rutube_video(url)
        elif 'twitch.tv' in url:
            parse_twitch_video(url)
        elif 'vimeo.com' in url:
            parse_vimeo_video(url)
        else:
            print(f"Unknown video service for URL: {url}")

def mkdir(path):
    if not os.path.exists(path):
        os.makedirs(path)

# Пример использования
urls = [
    'https://i.imgur.com/rBxB9ih.mp4',
    'https://packaged-media.redd.it/9kr7fm5k16eb1/pb/m2-res_480p.mp4?m=DASHPlaylist.mpd&v=1&e=1690448400&s=a56139280f0853d7bf356fe10e02cf0da081ba56#t=0',
    #'https://www.youtube.com/watch?v=dQw4w9WgXcQ',
    'https://giphy.com/gifs/nba-basketball-jeremy-lin-l0HlMAdOVOraymWvG'
]

download_videos(urls)
```
