```python
import wikipediaapi
import json

def fetch_wikimedia_data(subject):
    user_agent = 'MyWikiScraper/1.0'
    wiki_wiki = wikipediaapi.Wikipedia(
        language='en',
        extract_format=wikipediaapi.ExtractFormat.WIKI,
        user_agent=user_agent
    )
    page_py = wiki_wiki.page(subject)  # Получаем страницу по названию
    if not page_py.exists():
        print(f"Страница '{subject}' не найдена.")
        return
    
    main_content = page_py.text  # Основной контент узла (текст статьи)
    
    links = []
    for link_title in page_py.links:
        links.append(link_title)  # Список связанных страниц (гиперссылки)
    
    data = {
        'main_content': main_content,
        'links': links
    }
    
    return json.dumps(data)  # Преобразуем данные в JSON-структуру

# Пример использования
subject = 'Metallurgy'  # Запрос о предметной области
data = fetch_wikimedia_data(subject)
print(data)
```
