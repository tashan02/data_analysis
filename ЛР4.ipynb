```python

from sklearn.neighbors import NearestNeighbors

# Создание списка с рейтингами
movies_series = {
    'Твое сердце будет разбито': 8,
    'По осколкам твоего сердца': 6,
    'ЛюбовьНенависть': 7,
    'НенавистьЛюбовь': 8,
    'Поклонник': 8,
    'Влюбленная Ведьма': 9,
    'Восхитительная ведьма': 10,
    'Война и мир': 1,
    'Агрессор': 2,
    'Маленькие женщины': 10,
    'Полиана':10,
    'За руку с ветром':3,
    'Милые кости':4,
    'Худшие подруги':5,
    'КостяНика':1
}

# Передача списка другому человеку
another_person_movies_series = movies_series.copy()

# Добавление новых значений в список
another_person_movies_series['Именинница'] = 9
another_person_movies_series['Дурное поведение'] = 7
another_person_movies_series['Доверие'] = 1
another_person_movies_series['Магазин снов'] = 2
another_person_movies_series['Через год в это же время'] = 3
another_person_movies_series['Прекрасное несчатье'] = 4
another_person_movies_series['Уродливая любовь'] = 5
another_person_movies_series['Не ной'] = 7
another_person_movies_series['Не тупи'] = 6
another_person_movies_series['Бурундук'] = 8
another_person_movies_series['Не дружи со мной'] = 9
another_person_movies_series['Передружба'] = 10
another_person_movies_series['Не спасай меня'] = 1
another_person_movies_series['Влюбить за 90 секунд'] = 1
another_person_movies_series['Только попробуй уйти'] = 2
another_person_movies_series['Нелюбовь сероглазого короля'] = 5
another_person_movies_series['Любовь под напряжением'] = 5
another_person_movies_series['Поцелуй под омелой'] = 3
another_person_movies_series['Там, где живет любовь'] = 4
another_person_movies_series['Ты мое счастье'] = 4
another_person_movies_series['Любовь не по сценарию'] = 8
another_person_movies_series['Все из-за тебя'] = 9
another_person_movies_series['Зимняя любовь'] = 10
another_person_movies_series['Сказка о снежной принцессе'] = 10
another_person_movies_series['Загадай любовь'] = 10
another_person_movies_series['Самая белая ночь'] = 10
another_person_movies_series['Счастье в тебе'] = 10

# Создание матрицы признаков и списка рейтингов
X = []
y = []
for item, rating in another_person_movies_series.items():
    X.append([rating])
    y.append(item)

# Создание и обучение модели k-ближайших соседей
knn = NearestNeighbors(n_neighbors=3)
knn.fit(X)

# Определение элементарного рекомендательного алгоритма
new_movie_rating = 7  # Оценка нового значения
_, indices = knn.kneighbors([[new_movie_rating]])
recommended_movies = [y[i] for i in indices[0]]

# Передача рекомендации
print("Мой список:")
for item, rating in movies_series.items():
    print(f"{item}: {rating}")
    
print("Рекомендация:")
for movie in recommended_movies:
    print(movie)
```
